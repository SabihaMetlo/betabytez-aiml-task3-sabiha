```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("All libraries imported successfully!")
```

```python
# Load Dataset
df = pd.read_csv('model/train.csv')

print("\nFirst 5 rows")
df.head()
```

```python
# Basic Dataset Information
print("=== Dataset shape ===")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")

print("\n=== DataTypes ===")
print(df.info())
```

```python
# Fill Nan Values
fill_Nan_values = ["Alley", "PoolQC", "Fence", "MiscFeature", "FireplaceQu"]
df[fill_Nan_values] = df[fill_Nan_values].fillna("None")

print(df.info())
print(df.isnull().sum())
```

```python
df[fill_Nan_values] = df[fill_Nan_values].fillna("None")
df["GarageYrBlt"] = df["GarageYrBlt"].fillna(0)

pd.set_option('display.max_rows', None)
print(df.isnull().sum())
```

```python
# Step: Handle "NaN = No Feature" Columns
columns_none = ["Alley", "PoolQC", "Fence", "MiscFeature", "FireplaceQu"]

df[columns_none] = df[columns_none].fillna("None")

print(df.isnull().sum())
```

```python
# Handle "NaN = No Feature" — Basement, Garage, Masonry Columns
columns_none_group2 = [
    "BsmtQual", "BsmtCond", "BsmtExposure", "BsmtFinType1", "BsmtFinType2",
    "GarageType", "GarageFinish", "GarageQual", "GarageCond",
    "MasVnrType"
]

df[columns_none_group2] = df[columns_none_group2].fillna("None")

print(df.isnull().sum())
```

```python
# Handle Genuinely Missing Values

df["LotFrontage"] = df["LotFrontage"].fillna(df["LotFrontage"].median())
df["MasVnrArea"] = df["MasVnrArea"].fillna(0)
df["Electrical"] = df["Electrical"].fillna(df["Electrical"].mode()[0])

print(df.isnull().sum().sum())  # should print 0 if everything is handled
```

## Step: Explore Target Variable Distribution (SalePrice)

Before modeling, we check whether SalePrice is normally distributed or skewed.
A skewed target can bias regression models toward extreme values.

```python
plt.figure(figsize=(8,5))
sns.histplot(df["SalePrice"], kde=True)

plt.title("Distribution of SalePrice") 
plt.xlabel("Price")
plt.ylabel("Number of Houses")

plt.show()
```

```python
## Step: Log-Transform SalePrice to Reduce Skew
df["SalePrice_log"] = np.log1p(df["SalePrice"])

plt.figure(figsize=(8,5))
sns.histplot(df["SalePrice_log"], kde=True)
plt.title("Distribution of Log-Transformed SalePrice")
plt.xlabel("Log(SalePrice)")
plt.ylabel("Number of Houses")
plt.show()
```

```python
# Select only numeric columns for correlation analysis
numeric_df = df.select_dtypes(include=["int64", "float64"])

# Compute correlation matrix
corr_matrix = numeric_df.corr()

plt.figure(figsize=(16,12))
sns.heatmap(corr_matrix, cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

```python
# Correlation of every numeric feature with SalePrice, sorted highest to lowest
correlations = numeric_df.corr()["SalePrice"].sort_values(ascending=False)
print(correlations)
```

## Step: Scatter Plots — Top Correlated Features vs SalePrice

Correlation values tell us strength, but not shape.
Scatter plots let us visually confirm the relationship and spot outliers.

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: OverallQual vs SalePrice
sns.scatterplot(x=df["OverallQual"], y=df["SalePrice"], ax=axes[0])
axes[0].set_title("OverallQual vs SalePrice")

# Plot 2: GrLivArea vs SalePrice
sns.scatterplot(x=df["GrLivArea"], y=df["SalePrice"], ax=axes[1])
axes[1].set_title("GrLivArea vs SalePrice")

plt.tight_layout()
plt.show()
```

```python
# Step: Remove Known Outliers

# Check how many rows match this outlier condition before removing
outliers = df[(df["GrLivArea"] > 4000) & (df["SalePrice"] < 300000)]
print("Number of outliers found:", len(outliers))

# Remove these rows from the dataset
df = df[~((df["GrLivArea"] > 4000) & (df["SalePrice"] < 300000))]

print("New dataset shape:", df.shape)
```
